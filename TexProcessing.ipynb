## **Stopwords Removal, Word Tokenize and Sentence Tokenize**

```
import nltk
import re
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk import sent_tokenize 

sentence= "Modern technology is improving the current standard of education in the world. # used in twitter to highlight the events. 26 jan is our republic day."
text=' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",sentence).split())
print "sentence after punctuation removal", text

text = ''.join([i for i in text if not i.isdigit()])
print "sentence after digits removal", text

stop_words = set(stopwords.words('english')) 

tk = word_tokenize(text) 

print text
print "tokens", tk

filtered_sentence = [w for w in tk if not w in stop_words] 
print "----combined way----"
print filtered_sentence

print "-----the longer route----"

filtered_sentence = [] 

for w in tk: 
	if w not in stop_words: 
		filtered_sentence.append(w) 
print(filtered_sentence) 

ex="It is quite cold today. You might consider a jacket!"
ex1=sent_tokenize(ex)
print "----sentences----"
print(ex1)



```

# Synonyms and Antonym

```
from nltk.corpus import wordnet 

syns = wordnet.synsets("light") 
print syns

synonyms = [] 
antonyms = [] 
  
for syn in wordnet.synsets("light"): 
    for l in syn.lemmas(): 
        synonyms.append(l.name()) 
        if l.antonyms(): 
            antonyms.append(l.antonyms()[0].name()) 
  
print(set(synonyms)) 
print(set(antonyms)) 
```
